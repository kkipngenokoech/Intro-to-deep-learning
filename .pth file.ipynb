# PTH FILE

This is a checkpoint file used to save and load models, training progress, or any other python objects during and after training.

It is a binary file saved using pytorch `torch.save()` function.

## Types of PTH files

It can be used to store the following:

1. state dictionary - it contains weights and biases of the models. It doesn't include the model architecture. `torch.save(model.state_dict(), 'model.pth')`
2. Entire model - it saves both the model architecture and parameters, it can be loaded without redefining the architecture. `torch.save(model, 'model.pth')`
3. Training checkpoint, like model state dictionary, optimizer state dictionary, current epoch and best validation score:

```python
torch.save({
    'epoch': epoch,
    'model_state_dict': model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'loss': loss,
}, 'checkpoint.pth')
```
